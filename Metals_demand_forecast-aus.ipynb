```python
#import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DayLocator, HourLocator, DateFormatter, drange
import numpy as np
import dateutil.parser
from matplotlib import dates
from matplotlib import pyplot as plt
import datetime as dt
from datetime import timedelta


#pd.read_excel('path/to/file.xlsx', engine='openpyxl')
xls = pd.read_excel('Resources-and-Energy-Quarterly-September-2022-Forecast-Data.xls')

### Gold production forecast
df1 = pd.read_excel('Resources-and-Energy-Quarterly-September-2022-Forecast-Data.xls', sheet_name='Gold', skiprows=6)

columns_to_keep = [x for x in range(df1.shape[1]) if x not in [0,1,2,3,4,6, 11,12,13]]
df1 = df1.iloc[:, columns_to_keep]
df1 = df1.dropna(how='all')

df1.columns = df1.iloc[0] 
df1.columns = ['Gold', 2021, 2022, 2023, 2024]
df1.set_index('Gold', inplace=True)
df1.head(25)
```

```python
### Copper production forecast
df2 = pd.read_excel('Resources-and-Energy-Quarterly-September-2022-Forecast-Data.xls', sheet_name='Copper', skiprows=6)

columns_to_keep = [x for x in range(df2.shape[1]) if x not in [0,1,2,3,4,6,11,12,13,14, 15, 16, 17]]
df2 = df2.iloc[:, columns_to_keep]
df2 = df2.dropna(how='all')

df2.columns = df2.iloc[0] 
df2.columns = ['Copper', 2021, 2022, 2023, 2024]
df2.set_index('Copper', inplace=True)
df2.head(25)
```

```python
### Aliminium production forecast
df3 = pd.read_excel('Resources-and-Energy-Quarterly-September-2022-Forecast-Data.xls', sheet_name='Aluminium', skiprows=5)

columns_to_keep = [x for x in range(df3.shape[1]) if x not in [0,1,2,3,4,6,11,12,13,14]]
df3 = df3.iloc[:, columns_to_keep]
df3 = df3.dropna()

df3.columns = df3.iloc[0] 
df3.columns = ['Aluminium', 2021, 2022, 2023, 2024]
df3.set_index('Aluminium', inplace=True)
df3.head(25)
```

```python
### Nickel production forecast
df4 = pd.read_excel('Resources-and-Energy-Quarterly-September-2022-Forecast-Data.xls', sheet_name='Nickel', skiprows=5)

columns_to_keep = [x for x in range(df4.shape[1]) if x not in [0,1,2,3,4,6,11,12,13,14]]
df4 = df4.iloc[:, columns_to_keep]
df4 = df4.dropna(how='all')

df4.columns = df4.iloc[0] 
df4.columns = ['Nickel', 2021, 2022, 2023, 2024]
df4.set_index('Nickel', inplace=True)
#df4.replace('—', '')
df4.head(25)
```

```python
X = df4.columns


au_production = df1.iloc[4, :]
au_consumption =  df1.iloc[3, :]
cu_production = df2.iloc[4, :]
cu_consumption =  df2.iloc[5, :]
al_production = df3.iloc[0, :]
al_consumption =  df3.iloc[1, :]
ni_production = df4.iloc[4, :]
ni_consumption = df4.iloc[5, :]
# iloc[row slicing, column slicing]

fig, ax = plt.subplots(figsize = (10, 5))
#plt.title('Title')


#ax.plot(X, ni_production, color = 'b')

ax.plot(X, au_production, 'y--', label='Au production, [t]')
ax.plot(X, au_consumption, 'y',  label='Au consumption, [t]')

ax.plot(X, cu_production, 'r--', label='Cu production, [kt]')
ax.plot(X, cu_consumption, 'r',  label='Cu consumption, [kt]')

ax.plot(X, al_production, 'g--', label='Al production, [kt]')
ax.plot(X, al_consumption, 'g',  label='Al consumption, [kt]')

ax.plot(X, ni_production, 'k--', label='Ni production, [kt]')
ax.plot(X, ni_consumption, 'k',  label='Ni consumption, [kt]') #drawstyle='steps-post'

xticks=[2021, 2022, 2023, 2023, 2024, 2025]
ax.set_xticks(xticks)
plt.setp(ax.get_xticklabels(), fontsize=15)

ax.set_xlabel('Years', fontsize=18)
ax.set_ylabel('Production', fontsize=18)
plt.legend(fontsize=18)
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0), prop={'size':14}, fancybox=True, shadow=True)
  
plt.tight_layout()

plt.show()
```
